```python
#LIST: A list in Python is an ordered, mutable (changeable) collection that can store multiple items of different data types.

lst = [1, 2, 3, 'hello', 'c', [2,3,4]]
print(lst)

#append()
lst.append([3,'world'])
print(lst)

#extend() --->>> for seperate elements
lst.extend([2,3,4])
print(lst)


#insert() --->> add elements in between
#lst.insert(index,value)

lst.insert(2,"value")
print(lst)

#pop() --->> removes element from last
lst.pop()
print(lst)

#remove() --->> is used to remove the first occurrence of a specified value from a list.
#lst.remove(value)
lst.remove(2)
print(lst)

#index() --->> returns the position of the first occurrence of a specified value in a sequence
print(lst.index(2))

#count() ----->>> returns the number of times a specified value appears
print(lst.count(3))

#sort() --->>> sorts a list in place (modifies the original list)
lst2 = [5,6,7,1,2,3]
lst.sort()
print(lst2)

#sum() -->>>> used to add all the elements of an iterable (such as a list or tuple).


#min() -->> returns the smallest value from an iterable or from multiple values.

#max() --->> returns the largest value from an iterable or from multiple values.

#sorted() --->>> returns a new sorted list from an iterable without modifying the original object.
lst2 =[-1,2,8,120,6,100]
val = sorted(lst2)
print(val)






```

```python
# TUPLE: A tuple in Python is an ordered, immutable (cannot be changed after creation) collection of items.
tp = (1,1,1,2,53,4)
val = tuple(sorted(tp))
print(val)

# Functions that can be used in tuple:
#1.count()
#2.index()
#3.len()
#4.sum()
#5.max(),min()
#6.sorted()

print(tp.count(1))
print(tp.index(2))
print(len(tp))
print(sum(tp))
print(max(tp))
print(min(tp))
print(sorted(tp))


```
