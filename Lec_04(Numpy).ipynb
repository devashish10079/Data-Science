```python
# Where
# np.where(condition, value_if_true, value_if_false)
# a = np.random.randint(2,15,6)
# a = np.array([1,23,25,45,423,2])
a = np.array([[1, 2],
              [3, 4]])


# b = a = np.random.randint(2,18,6)
# b = np.array([1,2,3,4,4,5])
b = np.array([[5, 6],
              [7, 8]])

c = np.where(b > 3,b,a)
# print(a)
# print(b)
# print(c)

# Sort
d = np.sort(a)
print(d)

# Sum
e = np.sum(a)
print(e)

# Max
f = np.max(a)
print(f)

# Min
g = np.min(a)
print(g)

# Unique
h = np.unique(a)
print(h)

# Matmul
g = np.matmul(a,b)
print(g)
# Mean
i = np.mean(a)
print(i)

# Median
j = np.median(a)
print(j)

# Variance
k = np.var(a)
print(k)

```
