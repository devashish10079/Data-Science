**STRING**

```python
st =""
print(len(st))

# UPPER --->>> converts all lowercase letters in a string to uppercase.
st = "anand"
upper = st.upper()
print(upper)

#LOWER --->>> converts all uppercase letters in a string to lowercase.
st1 = "DEVASHISH"
lower = st1.lower()
print(lower)

# TITLE --->>> onverts the first letter of each word to uppercase and the remaining letters to lowercase.

st2 = "devashish room no 49 boys hostel"
tit = st2.title()
print(tit)

# CAPITALIZE --->>> converts the first character of a string to uppercase and the rest of the characters to lowercase.
st3 = "devashish room no 49 boys hostel"
cap = st3.capitalize()
print(cap)

# STRIP --->>> used to remove whitespace (or specified characters) from the beginning and end of a string.
st4 = "  devashish room no 49 boys hostel  "
strip = st4.strip()
print(strip)

# FIND --->>> searches for a substring and returns the index of its first occurrence.
st5 = "devashish room no 49 boys hostel"
find = st5.find("room")
print(find)

# INDEX --->> searches for a substring and returns the index of its first occurrence.
st6 = "devashish room no 49 boys hostel"
index = st6.index("room")
print(index)

# STARTSWITH --->>> checks whether a string starts with a specified prefix.
st7 = "devashish room no 49 boys hostel"
startswith = st7.startswith("devashish")
print(startswith)

# ENDSWITH --->>> checks whether a string ends with a specified suffix.
st8 = "devashish room no 49 boys hostel"
endswith = st8.endswith("hostel")
print(endswith)
```
