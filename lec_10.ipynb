```python
# BAR GRAPH

subjects = ['Maths','Physics','CS','English']
scores = [88,76,98,99]

plt.figure(figsize=(4,4))
plt.bar( subjects,scores,color ='red',edgecolor = 'black')
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot
hours = np.random.randint(1,10,50)
marks = hours * 8+ np.random.randint(1,20,50)
plt.scatter(hours,marks,s = 4)
plt.title('Hours vs Marks')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM
data = np.random.normal(60,10,500)
# data = np.random.randint(1,100,500)
plt.hist(data,bins = 20,edgecolor = 'black',color = 'purple')
plt.title('Histogram')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# PIE PLOT
subjects = ['Maths','Physics','CS','English']
scores = [20,20,50,10]
# plt.pie(scores)
plt.pie(scores,labels = subjects,autopct = '%1.3f%%',shadow = True,startangle= 0,explode=(0,0.1,0,0))
plt.title('Marks in Subjects')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('student_data.csv')
df.head()
```

```python
sns.lineplot(data = df,x = 'study_hours',y = 'exam_score')
plt.tight_layout()
plt.show()
#
```

```python
def grade (score):
  if score >= 90:
    return 'A'
  elif score >= 80:
    return 'B'
  elif score >= 70:
    return 'C'
  elif score >= 60:
    return 'D'
  else:
    return 'E'


df['grade'] = df['exam_score'].apply(grade)
df.head()
```

```python
sns.barplot(data = df,x = 'grade',y = 'exam_score',estimator= 'mean')
plt.tight_layout()
plt.show()
#
```
